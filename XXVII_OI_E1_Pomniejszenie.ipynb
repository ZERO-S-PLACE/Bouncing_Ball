```python
import sys
from io import StringIO

data = """15
619 610 2
10933 09995 3
65 56 2
21 01 2
09 01 1
01 01 1
10 01 1
9 9 1
12 10 1
22 12 2
9999 9999 3
999999 999999 3
1002111 0002222 1
20000 20000 1
20001 20000 1
555 333 1
0555 0551 3
0555 0333 4
"""

input_lines = data.split("\n")

def input():
    return input_lines.pop(0)

#Algorithm
# in exercise it is assumed that lengths are equal, so i skip comparison
def suffixAisSmallerThanB(A,B,i):
  while i<len(A):
    if int(A[i])==int(B[i]):
      i+=1
      continue
    elif int(B[i])>int(A[i]):
      return True
    else:
      return False
  return False

def getSuffix(A,i,changes):
  suffix=''
  i+=1
  while changes>0:
    if  i>len(A):
      print("!"+A +" " +B)
      break
    if A[i]=='9':
      if len(A)-i==changes:
        suffix+='8'
        changes-=1
      else:
        suffix+='9'
    else:
      changes-=1
      suffix+='9'
    i+=1
  return suffix+A[i:]


cases=int(input())
for c in range(cases):

  data=input().split()
  A=data[0] #bigger
  B=data[1]
  k=int(data[2])
  n=len(A) #size of number
  index=0
  changesInPrefix=0
  indexBest=-1
  changesInPrefixBest=0 #for last best solution
  c='' #connecting number
  #Check starting zeros
  while index<len(A):
    if not B[index]=='0':
      if changesInPrefix > k or n-index < k-changesInPrefix:
        break
      elif changesInPrefix == k:
        if suffixAisSmallerThanB(A,B,index):
          indexBest=index
          changesInPrefixBest=changesInPrefix
          c=A[index]
        break
      else:
        indexBest=index
        changesInPrefixBest=changesInPrefix
    if not A[index]==B[index]:
      changesInPrefix+=1
    index+=1



  if c=='':
    c=str(int(B[indexBest])-1)
  if A[indexBest]==c:
    if k-changesInPrefixBest == n-indexBest:
      if(int(c)-1>=0):
        c=str(int(c)-1)
        changesInPrefixBest+=1
      else:
        while True:
          indexBest-=1
          if indexBest<0:
            break
          if not B[indexBest]=='0':
            c=str(int(B[indexBest])-1)
            if  A[indexBest]==B[indexBest]:
              changesInPrefixBest+=1
            elif A[indexBest]==c:
              changesInPrefixBest-=1
            break
          if not A[indexBest]==B[indexBest]:
              changesInPrefixBest-=1
  else:
    changesInPrefixBest+=1

  if indexBest==-1  or n-indexBest < k-changesInPrefixBest:
    print("-1")
  else:
    suffix=getSuffix(A,indexBest,k-changesInPrefixBest)
    print(B[:indexBest]+c+suffix)








```
